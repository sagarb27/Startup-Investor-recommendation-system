## Jupyter notebook to extract list of investors from 'startup3.csv' and store them in 'investors.csv'

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import csv
import random
from urllib.request import urlopen, Request
import os
import re
import numpy as np
```

```python
def extractInvestorsList(investorString):
    investorsList = []
    i = 0
    investor = ""
    while(i<len(investorString)):
        if investorString[i] == '[':
            investor = ""
        elif investorString[i] == ']':
            investorsList.append(investor)
        else:
            investor = investor + investorString[i]
        i+=1
    return investorsList

def extractSingleInvestor(investorList):
    nameList = []
    for investor in investorList:
        name = investor.partition('|')[0]
        nameList.append(name)
    return nameList
```

```python
with open('D:\\cmpe256\\Team project\\startups3.csv','r') as file1:
    csvreader = csv.reader(file1)
    row = csvreader.__next__()
    finalList = []
    for row in csvreader:
        if len(row) >= 18:
            investorsList = extractInvestorsList(row[17])
            namesList = extractSingleInvestor(investorsList)
            #print(namesList)
            finalList = finalList + namesList

    namesArray = np.array(finalList)
    print(namesArray.shape)
    print(np.unique(namesArray).shape)
    with open('D:\\cmpe256\\Team project\\investors.csv','w',newline="") as file2:
        csvwriter = csv.writer(file2)
        csvwriter.writerow(['Name'])
        for investor in np.unique(namesArray):
            print(investor)
            csvwriter.writerow([investor])
    file2.close()
file1.close()
```
