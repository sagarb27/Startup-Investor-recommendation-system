## Following code reads the list of startups from seed-db.com and stores it in 'startups.csv'

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://www.seed-db.com/accelerators/viewall?acceleratorid=1011'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
with open('startups.csv','w',newline="") as startupfile:
    writer = csv.writer(startupfile)
    writer.writerow(['Name','State','Crunchbase link','Angel Link','Website','Start date','Exit Value','Unknown','Funding'])
    i=1
    while (i<len(soup.findAll('tr'))):
        one_tr_tag = soup.findAll('tr')[i]
        #print(i)
        
        #Get state
        state = ""
        buttons = one_tr_tag.findAll('button',class_='btn btn-success btn-xs disabled')
        for button in buttons:
            state = button.get_text()
        if state == "":
            buttons = one_tr_tag.findAll('button',class_='btn btn-danger btn-xs disabled')
            for button in buttons:
                state = button.get_text()
        
        #Get Angel link
        angel = ""
        buttons = one_tr_tag.findAll('button',class_='btn btn-default btn-xs hidethiswhensmall')
        if len(buttons) > 0:
            if 'https://angel.' in buttons[0].findAll('a')[0]['href']:
                angel = buttons[0].findAll('a')[0]['href']
        
        #Get name of startup
        name = one_tr_tag.findAll('a')[0].get_text()
        
        #Get Crunchbase link
        crunchbase = ''
        for button in buttons:
            if 'http://www.crunchbase.com' in button.findAll('a')[0]['href'] or 'https://www.crunchbase.com' in button.findAll('a')[0]['href']:
                crunchbase = button.findAll('a')[0]['href']

        #Get website link
        website = one_tr_tag.findAll('a')[1]['href']
        
        #Get date
        date = one_tr_tag.findAll('td')[3].get_text()
        
        #Get Exit value
        exit_val = one_tr_tag.findAll('td')[4].get_text()
        
        #Get the unknown single letter field
        unknown = one_tr_tag.findAll('td')[5].get_text()
        unknown = unknown.replace('\n','')

        #Get funding
        funding = one_tr_tag.findAll('td')[6].get_text()
        
        #Write to csv file
        writer.writerow([name,state,crunchbase,angel,website,date,exit_val,unknown,funding])
        i+=1
startupfile.close()
```
